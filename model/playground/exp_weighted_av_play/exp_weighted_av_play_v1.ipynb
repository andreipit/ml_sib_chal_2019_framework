```python
def exponentially_weighted_averages(df_col, B=0.8): # B = 0.1 -> like Y_pred = Y; B = 0.5 -> average values; B = 0.9 -> like straight line  
    Y = df_col.values.reshape(1,len(df_col))
    L = Y.shape[1]
    y_pred_prev = Y[0,0]
    Y_pred = np.zeros((1,L))
    for i in range(L): Y_pred[0,i] = B*y_pred_prev + (1-B)*Y[0,i]
    return pd.Series(Y_pred.reshape(-1))
exponentially_weighted_averages(df.col2).plot()
```

```python
col = 'col2'
coeff = 0.3

df[col+'_smooth'] = df[col].copy()
for i in range(1,df.shape[0]):
    df[col+'_smooth'].iat[i] = coeff*df[col][i-1] + (1-coeff)*df[col][i]
df[col+'_smooth'].plot()    
```
